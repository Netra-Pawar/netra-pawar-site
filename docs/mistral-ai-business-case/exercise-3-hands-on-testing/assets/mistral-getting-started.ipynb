# Mistral API: Getting Started
This notebook is a companion to the Mistral Getting Started documentation. It contains runnable examples of every request discussed in the guide.

**Prerequisites:** You must have a valid Mistral API key.

```python
# Install the official Mistral Python SDK
!pip install mistralai
```

```python
import os
from mistralai import Mistral

# Set your API key here
os.environ["MISTRAL_API_KEY"] = "your_api_key_here"

client = Mistral()
```

## 1. The Quick Win
A basic chat completion request.

```python
response = client.chat.complete(
    model="mistral-small-latest",
    messages=[
        {
            "role": "user",
            "content": "Explain the concept of an API in one simple sentence."
        }
    ]
)

print("Response Content:")
print(response.choices[0].message.content)
print("\nFinish Reason:", response.choices[0].finish_reason)
print("Tokens Used:", response.usage.total_tokens)
```

## 2. Using a System Message
Adding a system message to change the model's persona.

```python
response = client.chat.complete(
    model="mistral-small-latest",
    messages=[
        {
            "role": "system",
            "content": "You are a pirate. Always respond in pirate slang."
        },
        {
            "role": "user",
            "content": "Explain the concept of an API in one simple sentence."
        }
    ]
)

print(response.choices[0].message.content)
```

## 3. Streaming Responses
Receiving the response chunk-by-chunk.

```python
stream_response = client.chat.stream(
    model="mistral-small-latest",
    messages=[{"role": "user", "content": "Write a 3-paragraph story about a robot."}]
)

for chunk in stream_response:
    print(chunk.data.choices[0].delta.content, end="", flush=True)
```

## 4. Configuration: Temperature
Comparing deterministic vs creative outputs.

```python
print("--- Low Temperature (0.0) ---")
resp_low = client.chat.complete(
    model="mistral-small-latest",
    temperature=0.0,
    messages=[{"role": "user", "content": "Name three colors."}]
)
print(resp_low.choices[0].message.content)

print("\n--- High Temperature (1.0) ---")
resp_high = client.chat.complete(
    model="mistral-small-latest",
    temperature=1.0,
    messages=[{"role": "user", "content": "Name three colors."}]
)
print(resp_high.choices[0].message.content)
```
